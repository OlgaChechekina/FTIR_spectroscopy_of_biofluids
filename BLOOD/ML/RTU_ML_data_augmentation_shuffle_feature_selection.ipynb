```python
import numpy as np
from sklearn.base import TransformerMixin, BaseEstimator

class GlobalStandardScaler(BaseEstimator, TransformerMixin):
    def __init__(self):
        pass
    
    def fit(self, X, y=None):
        self.mean_ = np.mean(X)
        self.std_ = np.std(X)
        return self
    
    def transform(self, X, y=None):
        return (X - self.mean_) / self.std_
```

```python
#Expand dataset, Function also available in ChemUtils
def dataaugment(x, betashift = 0.05, slopeshift = 0.05,multishift = 0.05):
    #Shift of baseline
    #calculate arrays
    beta = np.random.random(size=(x.shape[0],1))*2*betashift-betashift
    slope = np.random.random(size=(x.shape[0],1))*2*slopeshift-slopeshift + 1
    #Calculate relative position
    axis = np.array(range(x.shape[1]))/float(x.shape[1])
    #Calculate offset to be added
    offset = slope*(axis) + beta - axis - slope/2. + 0.5

    #Multiplicative
    multi = np.random.random(size=(x.shape[0],1))*2*multishift-multishift + 1

    x = multi*x + offset

    return x

X_train = X.reset_index(drop=True)
shift = np.std(X_train)*0.1
X_train_aug = pd.DataFrame(np.repeat(X_train.to_numpy(), repeats=100, axis=0))
X_train_aug.index = np.arange(len(X_train_aug))
X_train_aug = dataaugment(X_train_aug, betashift=0.1, slopeshift=0.05, multishift=0.1)
X_train_aug = pd.DataFrame(X_train_aug)

y_train_aug = np.repeat(y.to_numpy(), repeats=100, axis=0)
y_train_aug = pd.DataFrame(y_train_aug)


print(X_train_aug.shape) 
print(y_train_aug.shape) 

#X_train_aug
```

```python
from sklearn.linear_model import LinearRegression, Lasso
from sklearn.svm import SVR
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.metrics import mean_squared_error
import pandas as pd
import numpy as np

# Create a dictionary with the regression methods to use
reg_methods = {
    "Linear Regression": LinearRegression(),
    "Lasso Regression": Lasso(),
    "Support Vector Regression": SVR(),
    "Decision Tree Regression": DecisionTreeRegressor(),
    "Random Forest Regression": RandomForestRegressor(),
    "Gradient Boosting Regression": GradientBoostingRegressor()
}

# Initialize result DataFrames
mse_results = pd.DataFrame(columns=list(reg_methods.keys()))
rmse_results = pd.DataFrame(columns=list(reg_methods.keys()))
percentage_rmse_results = pd.DataFrame(columns=list(reg_methods.keys()))

# Iterate over each target variable in y_train_aug
for y_col in y_train_aug.columns:
    if y_col not in selected_features_dict:
        continue

    selected_features = selected_features_dict[y_col]

    X_train_sub = X_train_aug[selected_features]
    y_train_sub = y_train_aug[y_col]

    mse_row = {}  # Store MSE values for the current target variable
    rmse_row = {}  # Store RMSE values for the current target variable
    percentage_rmse_row = {}  # Store %RMSE values for the current target variable

    # Iterate over each regression method
    for method_name, method in reg_methods.items():
        method.fit(X_train_sub, y_train_sub)
        y_pred = method.predict(X_train_sub)

        mse = mean_squared_error(y_train_sub, y_pred)
        rmse = np.sqrt(mse)
        percentage_rmse = (abs(rmse) / abs(y_train_sub.mean())) * 100  # Calculate %RMSE

        mse_row[method_name] = mse
        rmse_row[method_name] = rmse
        percentage_rmse_row[method_name] = percentage_rmse

    # Append MSE values to the mse_results DataFrame
    mse_results = mse_results.append(mse_row, ignore_index=True)

    # Append RMSE values to the rmse_results DataFrame
    rmse_results = rmse_results.append(rmse_row, ignore_index=True)

    # Append %RMSE values to the percentage_rmse_results DataFrame
    percentage_rmse_results = percentage_rmse_results.append(percentage_rmse_row, ignore_index=True)

# Display the MSE results table
#print("MSE Results:")
#print(mse_results)

# Display the RMSE results table
#print("\nRMSE Results:")
#rmse_results

# Display the %RMSE results table
#print("\n%RMSE Results:")
percentage_rmse_results
```

```python
import keras
from keras.models import Sequential
from keras.layers import Dense, Dropout, Flatten
from keras.layers import Conv1D, Reshape, GaussianNoise
from keras.callbacks import ReduceLROnPlateau
import tensorflow as tf

#Hyperparameters for the network
DENSE = 128
DROPOUT = 0.5
C1_K  = 8 #Number of kernels/feature extractors for first layer
C1_S  = 32 #Width of the convolutional mini networks
C2_K  = 16
C2_S  = 32
activation='relu'
input_dim = X_train.shape[1]

def make_model():
    model = Sequential()
    #Adding a bit of GaussianNoise also works as regularization
    model.add(GaussianNoise(0.05, input_shape=(input_dim,)))
    #First two is number of filter + kernel size
    model.add(Reshape((input_dim, 1) ))
    model.add(Conv1D(C1_K, (C1_S), activation=activation))
    model.add(Conv1D(C2_K, (C2_S), activation=activation))
    model.add(Flatten())
    model.add(Dropout(DROPOUT))
    model.add(Dense(DENSE, activation=activation))
    model.add(Dense(1, activation='linear'))

    model.compile(loss='mse', optimizer=tf.keras.optimizers.Adam(lr=0.01)) 

    return model
```
